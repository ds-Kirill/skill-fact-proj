```python
def request(url):
    '''создаёт юзер агента, отпарвляет гет запрос, возвращает объект BeautifulSoup html.parser'''
    user = fake_useragent.UserAgent().google
    header = {'user-agent' : user}
    
    try:        
        resp = requests.get(url, headers=header) #, timeout=( , )
        resp.encoding = 'utf-8'
        if resp.status_code != 404:
            return BeautifulSoup(resp.text, 'html.parser')
        else:
            return False
    
    except:
        return False

    
```

```python
def get_links_from_page(page):
    '''собираю все ссылки на страницы предложений автомобилей'''
    
    try:
        offers_links = page.find_all('a', class_="Link ListingItemTitle-module__link")
    except:
        return False
    links_for_parse = []
    for link in offers_links:
        links_for_parse.append(link.get('href') + '\n')
    return links_for_parse
```
